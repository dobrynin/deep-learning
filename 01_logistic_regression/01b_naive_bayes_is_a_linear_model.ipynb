### Naive Bayes Discriminator Is A Linear Model

**Traditional**
\\[
\frac{
    \prob{\text{S} = 1 \condbar W_1 = w_1, \ldots, W_k = w_k}
}{
    \prob{\text{S} = 0 \condbar W_1 = w_1, \ldots, W_k = w_k}
}
=
\frac{
    \prob{\text{S} = 1}
}{
    \prob{\text{S} = 0}
}
\frac{
    \prob{W_1 = w_1 \condbar \text{S} = 1}
}{
    \prob{W_1 = w_1 \condbar \text{S} = 0}
}
\cdots
\frac{
    \prob{W_k = w_k \condbar \text{S} = 1}
}{
    \prob{W_k = w_k \condbar \text{S} = 0}
}
\\]

**Exponentiated**
\\[
\frac{
    \prob{\text{S} = 1}
}{
    \prob{\text{S} = 0}
}
\left(
    \frac{
        \prob{W_1 = 1 \condbar \text{S} = 1}
    }{
        \prob{W_1 = 1 \condbar \text{S} = 0}
    }
\right)^{w_1}
\left(
    \frac{
        \prob{W_1 = 0 \condbar \text{S} = 1}
    }{
        \prob{W_1 = 0 \condbar \text{S} = 0}
    }
\right)^{1 - w_1}
\cdots
\left(
    \frac{
        \prob{W_k = 1 \condbar \text{S} = 1}
    }{
        \prob{W_k = 1 \condbar \text{S} = 0}
    }
\right)^{w_k}
\left(
    \frac{
        \prob{W_k = 0 \condbar \text{S} = 1}
    }{
        \prob{W_k = 0 \condbar \text{S} = 0}
    }
\right)^{1 - w_k}
\\]

**Logged**
\\[
\begin{align}
    &
    \left(
        \prob{\text{S} = 1}
        -
        \prob{\text{S} = 0}
    \right)
\\
    &+
    w_1
    \left(
        \prob{W_1 = 1 \condbar \text{S} = 1}
        -
        \prob{W_1 = 1 \condbar \text{S} = 0}
    \right)
    +
    (1 - w_1)
    \left(
        \prob{W_1 = 0 \condbar \text{S} = 1}
        -
        \prob{W_1 = 0 \condbar \text{S} = 0}
    \right)
\\
    &\cdots
\\
    &+
    w_k
    \left(
        \prob{W_k = 1 \condbar \text{S} = 1}
        -
        \prob{W_k = 1 \condbar \text{S} = 0}
    \right)
    +
    (1 - w_k)
    \left(
        \prob{W_k = 0 \condbar \text{S} = 1}
        -
        \prob{W_k = 0 \condbar \text{S} = 0}
    \right)
\end{align}
\\]

**Linear Formula**
\\[
\begin{align}
    &
    \left(
        \prob{\text{S} = 1}
        -
        \prob{\text{S} = 0}
    \right)
    +
    \sum_{i = 1}^k
        \left(
            \prob{W_i = 0 \condbar \text{S} = 1}
            -
            \prob{W_i = 0 \condbar \text{S} = 0}
        \right)
\\
    &+
    \sum_{i = 1}^k
    w_i
    \Big[
        \left(
            \prob{W_i = 1 \condbar \text{S} = 1}
            -
            \prob{W_i = 1 \condbar \text{S} = 0}
        \right)
        -
        \left(
            \prob{W_i = 0 \condbar \text{S} = 1}
            -
            \prob{W_i = 0 \condbar \text{S} = 0}
        \right)
    \Big]
\end{align}
\\]

**Coefficients**
\\[
\theta_0
=
    \left(
        \prob{\text{S} = 1}
        -
        \prob{\text{S} = 0}
    \right)
    +
    \sum_{i = 1}^k
        \left(
            \prob{W_i = 0 \condbar \text{S} = 1}
            -
            \prob{W_i = 0 \condbar \text{S} = 0}
        \right)
\\
\theta_i
=
    \left(
        \prob{W_i = 1 \condbar \text{S} = 1}
        -
        \prob{W_i = 1 \condbar \text{S} = 0}
    \right)
    -
    \left(
        \prob{W_i = 0 \condbar \text{S} = 1}
        -
        \prob{W_i = 0 \condbar \text{S} = 0}
    \right)
\\]

**Simplified Linear**

\\[
\log
\frac{
    \prob{\text{S} = 1}
}{
    \prob{\text{S} = 0}
}
=
\theta_0
+
\sum_{i=1}^k
\theta_i w_i
\\]

**Log Odds Ratio**

\\[
\theta_i
=
\log
    \frac{
        \prob{W_i = 1 \condbar \text{S} = 1}
        \big/
        \prob{W_i = 1 \condbar \text{S} = 0}
    }{
        \prob{W_i = 0 \condbar \text{S} = 1}
        \big/
        \prob{W_i = 0 \condbar \text{S} = 0}
    }
=
\log
    \frac{
        \prob{W_i = 1 \condbar \text{S} = 1}
        \big/
        \prob{W_i = 0 \condbar \text{S} = 1}
    }{
        \prob{W_i = 1 \condbar \text{S} = 0}
        \big/
        \prob{W_i = 0 \condbar \text{S} = 0}
    }
\\]
